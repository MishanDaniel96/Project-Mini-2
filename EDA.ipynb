```python
#10-- Sorghum Production (Kharif and Rabi) by Region

state_col = 'State Name'
kharif_col = 'KHARIF SORGHUM PRODUCTION (1000 tons)'
rabi_col = 'RABI SORGHUM PRODUCTION (1000 tons)'
# 1. Define Regional Mapping (using the same logic as the execution)
def get_region(state):
    state = state.strip()
    if state in ['Uttar Pradesh', 'Punjab', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir']:
        return 'North India'
    elif state in ['Maharashtra', 'Gujarat', 'Rajasthan']:
        return 'West India'
    elif state in ['West Bengal', 'Bihar', 'Odisha', 'Jharkhand', 'Assam', 'Sikkim']:
        return 'East India'
    elif state in ['Andhra Pradesh', 'Telangana', 'Karnataka', 'Tamil Nadu', 'Kerala']:
        return 'South India'
    elif state in ['Madhya Pradesh', 'Chhattisgarh']:
        return 'Central India'
    else:
        return 'Other/UT'
# 2. Apply the mapping to create a 'Region' column
df['Region'] = df[state_col].apply(get_region)
# 3. Handle the -1.0 anomaly (replace with 0 for production sums)
df_cleaned = df.copy()
df_cleaned[[kharif_col, rabi_col]] = df_cleaned[[kharif_col, rabi_col]].replace(-1.0, 0)
# 4. Group by Region and calculate the total production
sorghum_by_region = df_cleaned.groupby('Region')[[kharif_col, rabi_col]].sum().reset_index()
# 5. Sort the results
sorghum_by_region['Total Sorghum Production'] = sorghum_by_region[kharif_col] + sorghum_by_region[rabi_col]
sorghum_by_region_sorted = sorghum_by_region.sort_values(by='Total Sorghum Production', ascending=False)
# --- Plotting: Stacked Bar Chart ---
plot_df = sorghum_by_region_sorted.set_index('Region')[[kharif_col, rabi_col]]
plot_df.columns = ['Kharif Production', 'Rabi Production']
plt.figure(figsize=(10, 6))
plot_df.plot(
    kind='bar',
    stacked=True,
    color=['#FFC300', '#581845'],#Purple,Yellow
    ax=plt.gca())
plt.title('Total Sorghum Production by Region (Kharif vs. Rabi)', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sorghum Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Season')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
